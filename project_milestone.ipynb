```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/GSE40279_FINAL_with_real_ages.csv')
print(f"Dataset shape: {data.shape}")
```

```python
print("Age statistics:")
print(data['age'].describe())
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(data['age'], bins=30, edgecolor='black')
plt.xlabel('Age')
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
correlations = data.corr()['age'].drop('age').abs().sort_values(ascending=False)
plt.barh(range(10), correlations.head(10).values)
plt.title('Top 10 Correlated CpG Sites')
plt.show()
```
